# Data Visualization

```python
df.columns
```

```python
plt.figure(figsize=(7, 4))
plt.title('Count of Gender')
# sns.countplot(data=df, x=df['Outcome'], palette='Set2')
ax = sns.countplot(data=df, x=df['sex'], palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

From the above figure we can say that there are 662 females and 676 are males. It means 

```python
plt.figure(figsize=(7, 4))
plt.title('Frequency of age who buys Insurance')
# sns.countplot(data=df, x=df['Outcome'], palette='Set2')
sns.set(style="whitegrid") 
sns.distplot(x=df['age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

From above plot we can conclude that the frequency of insurance buyers are more withing the are range of 20 to 30.

```python
plt.figure(figsize=(7, 4))
plt.title('Charges paid according to gender')
# sns.countplot(data=df, x=df['Outcome'], palette='Set2')
sns.set(style="whitegrid") 
sns.stripplot(data=df, x=df['sex'], y=df['charges'], palette='Set2')
plt.xlabel('Gender')
plt.ylabel('Charges')
plt.show()
```

Above figure states that most of the charges are paid by male buyers as the plot of male are dense in the bottom and middle range.

```python
# lets see how many of them are smokers
plt.figure(figsize=(7, 4))
plt.title('Count of Smoking')
# sns.countplot(data=df, x=df['Outcome'], palette='Set2')
ax = sns.countplot(data=df, x=df['smoker'], hue='sex', palette='Set1')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Smoke')
plt.ylabel('Count')
plt.show()
```

This graph shows that people who buys insurance and smoke are mostly on the other who buys insurance and does not smoke most of them 
are women.

```python
# lets check the region of people who buys insurance.
plt.figure(figsize=(9, 7))
plt.title('Region of Buyers')
# sns.countplot(data=df, x=df['Outcome'], palette='Set2')
ax = sns.countplot(data=df, x=df['region'], hue='sex', palette='Set3')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
# lets check the region of people who buys insurance.
plt.figure(figsize=(8, 5))
plt.title('Children of Insurance Buyers')
# sns.countplot(data=df, x=df['Outcome'], palette='Set2')
ax = sns.countplot(data=df, x=df['children'], palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Number of children')
plt.ylabel('Count')
plt.show()
```

Above graph shows that only 18 peoples have 5 children and most of the buyers doesnt have children.
